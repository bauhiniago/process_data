```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

print(x)
plt.plot(x, y1)
plt.plot(x, y2)
 
plt.title('line chart')
plt.xlabel('x')
plt.ylabel('y')
 
plt.show()
```
